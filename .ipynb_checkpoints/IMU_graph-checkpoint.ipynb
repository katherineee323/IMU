```python
import matplotlib.pyplot as plt
```

```python
Acc = []
Gyro = []
Ang = []
Mag = []
linecount = 0
with open("IMU.txt",'r') as file:
    for lines in file.readlines():
        if linecount % 4 == 0:
            Acc.append([float(x) for x in lines.split()])
        if linecount % 4 == 1:
            Gyro.append([float(x) for x in lines.split()])
        if linecount % 4 == 2:
            Ang.append([float(x) for x in lines.split()])
        if linecount % 4 == 3:
            Mag.append([float(x) for x in lines.split()])
        linecount += 1
```

```python
print(Acc)
```

```python
Acc_x = []
Acc_y = []
Acc_z = []
Gyro_x = []
Gyro_y = []
Gyro_z = []
Ang_x = []
Ang_y = []
Ang_z = []
Mag_x = []
Mag_y = []
Mag_z = []
for i in range (len(Acc)):
    Acc_x.append(Acc[i][0])
    Acc_y.append(Acc[i][1])
    Acc_z.append(Acc[i][2])
    Gyro_x.append(Gyro[i][0])
    Gyro_y.append(Gyro[i][1])
    Gyro_z.append(Gyro[i][2])
    Ang_x.append(Ang[i][0])
    Ang_y.append(Ang[i][1])
    Ang_z.append(Ang[i][2])
    Mag_x.append(Mag[i][0])
    Mag_y.append(Mag[i][1])
    Mag_z.append(Mag[i][2])
print(Acc_x)
```

```python
x = [i for i in range(len(Acc))]
print(x)
```

```python
plt.title("Acceleration graph")
plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.plot(x, Acc_x, 'green')
plt.plot(x, Acc_y, 'blue')
plt.plot(x, Acc_z, 'red')
plt.show()
```

```python
start = 4200
end = 4500
x2 = [i for i in range(start,end)]
Acc_x2 = Acc_x[start:end]
Acc_y2 = Acc_y[start:end]
Acc_z2 = Acc_z[start:end]
plt.title("Acceleration graph")
plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.plot(x2, Acc_x2, 'green')
plt.plot(x2, Acc_y2, 'blue')
plt.plot(x2, Acc_z2, 'red')
plt.show()
```

```python
plt.title("Gyroscope graph")
plt.xlabel("Time")
plt.ylabel("Gyro")
plt.plot(x, Gyro_x, 'pink')
plt.plot(x, Gyro_y, 'purple')
plt.plot(x, Gyro_z, 'orange')
plt.show()
```

```python
x2 = [i for i in range(start,end)]
Gyro_x2 = Gyro_x[start:end]
Gyro_y2 = Gyro_y[start:end]
Gyro_z2 = Gyro_z[start:end]
plt.title("Gyroscope graph")
plt.xlabel("Time")
plt.ylabel("Gyroscope")
plt.plot(x2, Gyro_x2, 'pink')
plt.plot(x2, Gyro_y2, 'purple')
plt.plot(x2, Gyro_z2, 'orange')
plt.show()
```

```python
plt.title("Angular graph")
plt.xlabel("Time")
plt.ylabel("Angle")
plt.plot(x, Ang_x, 'turquoise')
plt.plot(x, Ang_y, 'black')
plt.plot(x, Ang_z, 'gray')
plt.show()
```
